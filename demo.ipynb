```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
im = cv2.imread(f'harbor_images/Picture5.png')
```

```python
rgb_im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_im)
plt.title('rgb')
plt.show()
```

```python
hsv_im = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv_im)
plt.title('hsv')
plt.show()
```

```python
lab_im = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
```

```python
plt.imshow(lab_im)
plt.title('lab')
plt.show()
```

```python
plt.imshow(lab_im[:,:, 0])
plt.title('l channel')
plt.show()
```

```python
plt.imshow(lab_im[:,:, 1])
plt.title('Ca channel')
plt.show()
```

```python
plt.imshow(lab_im[:,:, 2])
plt.title('Cb channel')
plt.show()
```
